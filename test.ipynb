```python
# Stream a large CSV file
import csv
import urllib.request 
import codecs 
import numpy as np 

url = "https://data.heatonresearch.com/data/t81-558/iris.csv"
urlstream = urllib.request.urlopen(url)
csvfile = csv.reader(codecs.iterdecode(urlstream, 'utf-8'))
next(csvfile) # skip header row
sum = np.zeros(4)
count = 0

for line in csvfile:
    #Convert each row to numpy array
    line2 = np.array(line)[0:4].astype(float)

    #if the line is of right length (skip empty lines), then add
    if len (line2) == 4:
        sum += line2
        count += 1

    #calculate the average and print the average of the 4 iris measurements
    print(sum/count)
```
